```python

# Download stock data
df = yf.download("RELIANCE.NS", start="2023-01-01", end="2025-01-01")

print(df.head())

# Save CSV
df.to_csv("reliance_stock.csv")
```

```python
Rel = pd.read_csv("reliance_stock.csv")
Reliance_data = pd.DataFrame(Rel)
Reliance_data.head()
```

```python
Reliance_data.info()
```

```python

import numpy as np
Reliance_data.rename(columns={'Price':'date'},inplace = True)
Reliance_data.replace('Date',np.NaN,inplace = True)
Reliance_data.head()
```

```python
Reliance_data = Reliance_data.drop([0,1])
Reliance_data.reset_index(drop=True,inplace=True)
Reliance_data.head()
```

```python
numeric_cols = ['Close','High','Low','Open','Volume']
Reliance_data[numeric_cols] = Reliance_data[numeric_cols].astype('float')
Reliance_data['Date'] = pd.to_datetime(Reliance_data['Close'])
Reliance_data.head()
```

```python
Reliance_data.describe()
```

```python
Reliance_data.corr()
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
X = Reliance_data[['High','Low','Open','Volume']]
y = Reliance_data['Close']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,train_size=0.8, random_state=100)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
yhat = lm.predict(X)
yhat
print(lm.intercept_)
print(lm.coef_) #slope
```

```python
y_pred=lm.predict(X_test)
y_pred
```

```python
print('r2_score:',r2_score(y_test,y_pred))
print('mean_squared_error:',mean_squared_error(y_test,y_pred))
```

R2_score is close to 1,therefore model is a good fit.
MSE = 28.87, the square root of this will be ~ 5.3(approx) on average share price variation.


```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Share Price")
plt.show()
```
